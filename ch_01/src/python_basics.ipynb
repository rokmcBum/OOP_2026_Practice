## 1. Variables and Data Types

Python is dynamically typed — you don't need to declare a variable's type explicitly.

```python
# Integer
age = 25
print(f"age = {age}, type = {type(age)}")

# Float
pi = 3.14159
print(f"pi = {pi}, type = {type(pi)}")

# String
name = "Alice"
print(f"name = {name}, type = {type(name)}")

# Boolean
is_student = True
print(f"is_student = {is_student}, type = {type(is_student)}")

# NoneType
nothing = None
print(f"nothing = {nothing}, type = {type(nothing)}")
```

```python
# Type conversion
x = "42"
print(int(x) + 8)       # str -> int
print(float(x) + 0.5)   # str -> float
print(str(100) + " points")  # int -> str
```

## 2. Operators

```python
# Arithmetic operators
a, b = 17, 5
print(f"a + b  = {a + b}")
print(f"a - b  = {a - b}")
print(f"a * b  = {a * b}")
print(f"a / b  = {a / b}")      # true division (float)
print(f"a // b = {a // b}")     # floor division (int)
print(f"a % b  = {a % b}")      # modulo
print(f"a ** b = {a ** b}")     # exponentiation
```

```python
# Comparison and logical operators
x, y = 10, 20

print(f"x == y : {x == y}")
print(f"x != y : {x != y}")
print(f"x < y  : {x < y}")
print(f"x >= y : {x >= y}")

print(f"x > 5 and y < 30 : {x > 5 and y < 30}")
print(f"x > 15 or y < 30 : {x > 15 or y < 30}")
print(f"not (x == y)      : {not (x == y)}")
```

## 3. Strings

```python
# String basics
greeting = "Hello, World!"

print(greeting[0])        # indexing
print(greeting[7:12])     # slicing
print(greeting[::-1])     # reverse
print(len(greeting))      # length
```

```python
# Useful string methods
text = "  Object-Oriented Programming  "

print(text.strip())             # remove whitespace
print(text.strip().lower())     # lowercase
print(text.strip().upper())     # uppercase
print(text.strip().split("-"))  # split into list
print("_".join(["OOP", "2026", "Practice"]))  # join list into string

# f-string formatting
course = "OOP"
year = 2026
print(f"Welcome to {course} {year}!")
```

## 4. Lists, Tuples, and Dictionaries

```python
# List — mutable, ordered
fruits = ["apple", "banana", "cherry"]
fruits.append("date")
fruits.insert(1, "avocado")
print(f"fruits        : {fruits}")
print(f"fruits[1:3]   : {fruits[1:3]}")
print(f"len(fruits)   : {len(fruits)}")

removed = fruits.pop()  # removes last element
print(f"popped        : {removed}")
print(f"fruits after   : {fruits}")
```

```python
# Tuple — immutable, ordered
coordinates = (3, 4)
print(f"coordinates : {coordinates}")
print(f"x = {coordinates[0]}, y = {coordinates[1]}")

# Unpacking
x, y = coordinates
print(f"unpacked: x={x}, y={y}")

# Tuples are immutable — the following would raise an error:
# coordinates[0] = 10  # TypeError!
```

```python
# Dictionary — mutable, key-value pairs
student = {
    "name": "Alice",
    "age": 21,
    "major": "Computer Science",
}

print(f"name  : {student['name']}")
print(f"keys  : {list(student.keys())}")
print(f"values: {list(student.values())}")

# Add / update
student["gpa"] = 3.8
student["age"] = 22
print(f"updated: {student}")

# Safe access with .get()
print(f"phone: {student.get('phone', 'N/A')}")
```

```python
# List comprehension
squares = [x ** 2 for x in range(1, 11)]
print(f"squares: {squares}")

# Filtered comprehension
evens = [x for x in range(1, 21) if x % 2 == 0]
print(f"evens  : {evens}")

# Dict comprehension
square_dict = {x: x ** 2 for x in range(1, 6)}
print(f"square_dict: {square_dict}")
```

## 5. Control Flow

```python
# if / elif / else
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print(f"Score {score} -> Grade {grade}")
```

```python
# for loop
print("--- for loop ---")
languages = ["Python", "Java", "C++", "JavaScript"]
for lang in languages:
    print(f"  {lang} has {len(lang)} characters")

# for with enumerate
print("\n--- enumerate ---")
for i, lang in enumerate(languages):
    print(f"  [{i}] {lang}")

# for with range
print("\n--- range ---")
total = 0
for i in range(1, 101):
    total += i
print(f"  Sum of 1..100 = {total}")
```

```python
# while loop
print("--- while loop ---")
count = 5
while count > 0:
    print(f"  countdown: {count}")
    count -= 1
print("  Go!")

# break and continue
print("\n--- break & continue ---")
for n in range(1, 20):
    if n % 3 == 0:
        continue   # skip multiples of 3
    if n > 10:
        break      # stop at 10
    print(f"  {n}", end=" ")
print()
```

## 6. Functions

```python
# Basic function
def greet(name: str) -> str:
    """Return a greeting message."""
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet("Bob"))
```

```python
# Default arguments and keyword arguments
def power(base: float, exponent: float = 2) -> float:
    """Calculate base raised to the exponent (default: squared)."""
    return base ** exponent

print(f"power(3)      = {power(3)}")
print(f"power(2, 10)  = {power(2, 10)}")
print(f"power(5, exponent=3) = {power(5, exponent=3)}")
```

```python
# *args and **kwargs
def summarize(*args, **kwargs):
    """Demonstrate variable-length arguments."""
    print(f"  args   = {args}")
    print(f"  kwargs = {kwargs}")

summarize(1, 2, 3, name="Alice", age=21)
```

```python
# Lambda functions
double = lambda x: x * 2
print(f"double(7) = {double(7)}")

# Using lambda with built-in functions
words = ["banana", "apple", "cherry", "date"]
sorted_words = sorted(words, key=lambda w: len(w))
print(f"sorted by length: {sorted_words}")
```

## 7. File I/O Basics

```python
# Writing to a file
with open("sample.txt", "w") as f:
    f.write("Line 1: Hello, File I/O!\n")
    f.write("Line 2: Python makes it easy.\n")
    f.write("Line 3: Always use 'with' for safe file handling.\n")

print("File written successfully.")
```

```python
# Reading from a file
with open("sample.txt", "r") as f:
    content = f.read()
print(content)

# Reading line by line
with open("sample.txt", "r") as f:
    for i, line in enumerate(f, start=1):
        print(f"  [{i}] {line.strip()}")
```

```python
# Clean up the temporary file
import os
os.remove("sample.txt")
print("Temporary file removed.")
```
