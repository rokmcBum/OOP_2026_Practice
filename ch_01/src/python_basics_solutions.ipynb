## Exercise 1: FizzBuzz

Print numbers from 1 to 30. For multiples of 3, print `"Fizz"`. For multiples of 5, print `"Buzz"`. For multiples of both, print `"FizzBuzz"`.

```python
for n in range(1, 31):
    if n % 3 == 0 and n % 5 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)
```

## Exercise 2: Word Counter

Write a function `count_words(sentence: str) -> dict` that takes a sentence and returns a dictionary mapping each word (lowercased) to its frequency.

```python
def count_words(sentence: str) -> dict:
    counts = {}
    for word in sentence.lower().split():
        counts[word] = counts.get(word, 0) + 1
    return counts


# Test
print(count_words("the cat sat on the mat the cat"))
print(count_words("Hello hello HELLO world"))
```

## Exercise 3: List Flattening

Write a function `flatten(nested: list) -> list` that takes a nested list (e.g., `[[1, 2], [3, [4, 5]]]`) and returns a flat list `[1, 2, 3, 4, 5]`.

```python
def flatten(nested: list) -> list:
    result = []
    for item in nested:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result


# Test
print(flatten([[1, 2], [3, [4, 5]]]))
print(flatten([1, [2, [3, [4, [5]]]]]))
print(flatten([[1], 2, [[3]], [4, [5, 6]]]))
```
